```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.shape
```

```python
iris.dtypes
```

```python
sns.distplot(iris['sepal_length'])
```

# kde :bool, optional
- Whether to plot a gaussian kernel density estimate.
- Eliminates the distribution function kde represents Kernel Density Estimation

```python
sns.distplot(iris['sepal_length'],kde = False)
```

# hist :bool, optional
- Whether to plot a (normed) histogram.
- Elimating the histogram using hist -> False by default it will be True

```python
sns.distplot(iris['sepal_length'],hist= False)
```

# rug : bool, option
- Whether to draw a rugplot on the support axis
- rug gives the value within which the entire data is located

```python
sns.distplot(iris['sepal_length'],kde = False,rug = True)
```

# hist_kws : dict, optional
- Keyword arguments for matplotlib.axes.Axes.hist().

```python
sns.distplot(iris['sepal_length'],kde = False, 
             hist_kws = {"histtype": "step", "linewidth": 3,
                            "alpha": 1, "color": "y"})
```

# kde_kws : dict, optional
- Keyword arguments for kdeplot().
- seaborn.kdeplot(data, data2=None, shade=False, vertical=False, kernel='gau', bw='scott', gridsize=100, cut=3, clip=None, legend=True, cumulative=False, shade_lowest=True, cbar=False, cbar_ax=None, cbar_kws=None, ax=None, **kwargs)

```python
sns.distplot(iris['sepal_length'],hist= False,kde_kws = {"color": "c", "lw": 3, "label": "KDE",'shade':True})
```

# rug_kws : dict, optional
- Keyword arguments for rugplot().

```python
sns.distplot(iris['sepal_length'],kde = False,rug = True,rug_kws = {'color': 'r','label' : 'Rug'})
```
