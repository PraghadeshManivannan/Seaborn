```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
sns.violinplot(x="petal_length", y="species", data=iris,palette='rainbow')
```

```python
sns.violinplot(x="species", y="sepal_length", data=iris,palette = 'rainbow')
```

```python
sns.violinplot(x="species", y="sepal_length", data=iris,palette = 'rainbow',inner = 'stick')
```

```python
sns.violinplot(x="species", y="sepal_length", data=iris,palette = 'rainbow',inner = 'point')
```

```python
sns.violinplot(x="species", y="sepal_length", data=iris,palette = 'rainbow',inner = 'quartile')
```

```python
sns.violinplot(x="species", y="sepal_length", data=iris,palette = 'rainbow',inner = None)
```
