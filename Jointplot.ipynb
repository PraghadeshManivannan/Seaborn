```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

# seaborn.jointplot
##### seaborn.jointplot(x, y, data=None, kind='scatter', stat_func=None, color=None, height=6, ratio=5, space=0.2, dropna=True, xlim=None, ylim=None, joint_kws=None, marginal_kws=None, annot_kws=None, **kwargs)
- x, y : Data or names of variables in data.
- data : DataFrame when x and y are variable names.

```python
sns.jointplot(x='petal_length',y = 'sepal_length',data = iris)
```

# Kind

#### kind -> { “scatter” | “reg” | “resid” | “kde” | “hex” }, optional
- Kind of plot to draw.
- By default it will be kind -> scatter

```python
sns.jointplot(x='petal_length',y = 'sepal_length',data = iris,kind = 'reg')
```

```python
sns.jointplot(x='petal_length',y = 'sepal_length',data = iris,kind = 'resid')
```

```python
sns.jointplot(x='petal_length',y = 'sepal_length',data = iris,kind =  'kde')
```

```python
sns.jointplot(x='petal_length',y = 'sepal_length',data = iris,kind =  'hex')
```

```python
sns.jointplot(x='petal_length',y = 'sepal_length',data = iris,kind =  'kde',color = 'c',space = 0)
```
